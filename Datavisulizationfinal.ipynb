```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df1 = pd.read_csv('argentina.csv')
```

```python
df2 = pd.read_csv('france.csv')
```

```python
df1['TARGET'] = df1['TARGET'].replace({-1: 'Lost', 0: 'Draw', 1: 'Win'})
df2['TARGET'] = df2['TARGET'].replace({-1: 'Lost', 0: 'Draw', 1: 'Win'})
```

```python
sns.countplot(data=df1, x='TARGET')
plt.xlabel('Match Outcome')
plt.ylabel('Count')
plt.title('Distribution of Match Outcomes for argentina')
plt.show()
```

```python
sns.countplot(data=df2, x='TARGET')
plt.xlabel('Match Outcome')
plt.ylabel('Count')
plt.title('Distribution of Match Outcomes for france')
plt.show()
```

```python
sns.pairplot(df1[['GD', 'ATTD', 'MIDD', 'DEFD']], diag_kind='kde')
plt.suptitle('Pairplot of Numeric Attributes for argenina', y=1.02)
plt.show()
```

```python
sns.pairplot(df2[['GD', 'ATTD', 'MIDD', 'DEFD']], diag_kind='kde')
plt.suptitle('Pairplot of Numeric Attribufrancetes for francxe', y=1.02)
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
sns.boxplot(data=df1, x='TARGET', y='MIDD', ax=axes[0])
sns.boxplot(data=df1, x='TARGET', y='ATTD', ax=axes[1])
sns.boxplot(data=df1, x='TARGET', y='DEFD', ax=axes[2])
plt.suptitle('Numeric Attributes by Match Outcome for argentina', y=1.02)
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
sns.boxplot(data=df2, x='TARGET', y='MIDD', ax=axes[0])
sns.boxplot(data=df2, x='TARGET', y='ATTD', ax=axes[1])
sns.boxplot(data=df2, x='TARGET', y='DEFD', ax=axes[2])
plt.suptitle('Numeric Attributes by Match Outcome for france', y=1.02)
plt.show()
```
